```python
import cv2
import mediapipe as mp
import os
import multiprocessing
import numpy as np
```

```python
# Mediapipe setup
mp_face_detection = mp.solutions.face_detection
mp_drawing = mp.solutions.drawing_utils

# Input/output directories
input_dir = "/content/deepfake-detection/dataset/Celeb-DF-v2/Celeb-synthesis"
output_dir = "/content/deepfake-detection/processed_faces/original"
os.makedirs(output_dir, exist_ok=True)
```

```python

def process_video(video_path, serial_no, max_frames=200, sampling="uniform"):
    """
    Extract up to `max_frames` faces from a video using MediaPipe.
    sampling: "uniform" (evenly spread across duration) or "random".
    Saves cropped faces into <output_dir>/<video_name>/.
    """
    video_name = os.path.splitext(os.path.basename(video_path))[0]
    save_dir = os.path.join(output_dir, video_name)

    # ✅ Skip if already processed with enough frames
    if os.path.exists(save_dir):
        existing = [f for f in os.listdir(save_dir) if f.endswith(".jpg")]
        if len(existing) >= max_frames:
            print(f"{serial_no}. ⏩ Skipping {video_name} (already has {len(existing)} frames)")
            return

    os.makedirs(save_dir, exist_ok=True)

    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        print(f"{serial_no}. ❌ Cannot open: {video_path}")
        return

    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT) or 0)

    if total_frames > 0:
        n = min(max_frames, total_frames)
        if sampling == "uniform":
            indices = np.linspace(0, total_frames - 1, num=n, dtype=int)
        else:
            rng = np.random.default_rng()
            indices = np.sort(rng.choice(total_frames, size=n, replace=False))
    else:
        indices = list(range(max_frames))

    saved = 0
    with mp_face_detection.FaceDetection(model_selection=1, min_detection_confidence=0.5) as face_det:
        for idx in indices:
            if total_frames > 0:
                cap.set(cv2.CAP_PROP_POS_FRAMES, int(idx))

            ret, frame = cap.read()
            if not ret or frame is None:
                continue

            results = face_det.process(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
            if not results.detections:
                continue

            h, w, _ = frame.shape
            for det_i, det in enumerate(results.detections):
                box = det.location_data.relative_bounding_box
                x = max(0, int(box.xmin * w))
                y = max(0, int(box.ymin * h))
                bw = max(1, int(box.width * w))
                bh = max(1, int(box.height * h))

                face = frame[y:y+bh, x:x+bw]
                if face.size == 0:
                    continue

                out_path = os.path.join(save_dir, f"{video_name}_f{idx}_d{det_i}.jpg")
                cv2.imwrite(out_path, face)

            saved += 1
            if saved >= max_frames:
                break

    cap.release()
    print(f"{serial_no}. ✅ Done: {video_name} — sampled {saved} frames ({sampling})")


# Collect all .mp4 files

all_videos = []
for root, dirs, files in os.walk(input_dir):
    for f in files:
        if f.endswith(".mp4"):
            all_videos.append(os.path.join(root, f))
            if len(all_videos) == 100:  # Stop once we have 100 videos
                break
    if len(all_videos) == 100:
        break

print(f"Total videos: {len(all_videos)}")

# Pair serial numbers with videos
video_with_ids = [(vp, i) for i, vp in enumerate(all_videos, start=1)]

# Run processing (parallel across videos)
if __name__ == "__main__":
    num_workers = 2  # Adjust depending on Colab runtime
    with multiprocessing.Pool(num_workers) as pool:
        pool.starmap(process_video, video_with_ids)
```
